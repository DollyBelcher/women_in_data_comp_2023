```python
import pandas as pd
```

```python
grant_df = pd.read_csv("../Data/Community_Development_Block_Grant_Activity.csv")
carbon_df = pd.read_csv("../Data/carbon_years.csv")
states_df = pd.read_csv("../Data/states.csv")
```

```python
states_dict = dict(zip(states_df["State"], states_df["Postal"]))
```

```python
carbon_df["State"].replace(states_dict, inplace=True)
```

```python
full_df = carbon_df.merge(grant_df, left_on="State", right_on="STATE")
full_df = full_df[["STATE", "GROUPING", "ACTV_FUNDING_AMT", "2021"]]
```

```python
result = full_df.groupby('STATE').agg({
    'ACTV_FUNDING_AMT': 'sum',
    '2021': 'mean'
}).reset_index()
```

```python
result["ACTV_FUNDING_AMT"].corr(result["2021"])
```

```python
result["ACTV_FUNDING_AMT"].corr(result["2021"])
```

```python
# Group by 'STATE' and 'GROUPING', then aggregate
grouped = full_df.groupby(['STATE', 'GROUPING']).agg({
    'ACTV_FUNDING_AMT': 'sum',
    '2021': 'mean'
}).reset_index()

# Get rows with the highest 'ACTV_FUNDING_AMT' for each 'STATE'
idx = grouped.groupby('STATE')['ACTV_FUNDING_AMT'].idxmin()
highest_funding_per_state = grouped.loc[idx]
```

```python
highest_funding_per_state.sort_values("2021").reset_index()
```
